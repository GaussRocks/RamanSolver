# **Equação Raman**

Primeiro, considere a seguinte substituiação: 

$$
\begin{aligned}
&\theta(\Delta f) = \frac{g_R(\Delta f)}{\Gamma A_{eff}} \\
&\phi(\Delta f) = \left [ 1 + \left (  e^{\frac{h \Delta f}{\mathrm{K}\mathrm{T}}}-1\right)^{-1} \right]
\end{aligned}
$$

## Forward

$$

\begin{aligned}

&\begin{aligned}

\frac{\mathrm{d}P^{+}_i}{\mathrm{dz}} = &-\alpha_i P^{+}_i + \eta_i P^{-}_i - P^{+}_i \sum_{m=i+1}^{n} \frac{f_i}{f_m}\frac{g_R(f_i-f_m)}{\Gamma A_{eff}} [ P^{+}_i + P^{-}_i] +\\

&-2hf_i P^{+}_i \sum_{m=i+1}^{n} \frac{f_i}{f_m} \frac{g_R(f_i-f_m)}{\Gamma A_{eff}} \left [ 1 + \left (  e^{\frac{h(f_i-f_m)}{\mathrm{K}\mathrm{T}}}-1\right)^{-1} \right] \Delta v +\\ 

&+ P^{+}_i \sum_{m=0}^{i-1} \frac{g_R(f_m-f_i)}{\Gamma A_{eff}} [P^{+}_i + P^{-}_i] + \\

&+ hf_i\sum_{m=0}^{i-1} \frac{g_R(f_m-f_i)}{\Gamma A_{eff}} [ P^{+}_i + P^{-}_i] \left [ 1 + \left (  e^{\frac{h(f_m-f_i)}{\mathrm{K}\mathrm{T}}}-1\right)^{-1} \right] \Delta v \\ 

\end{aligned}\\\\

&\begin{aligned}

\frac{\mathrm{d}P^{+}_i}{\mathrm{dz}} = &-\alpha_i P^{+}_i + \eta_i P^{-}_i - P^{+}_i \sum_{m=i+1}^{n} \frac{f_i}{f_m} \theta(\Delta f) [ P^{+}_i + P^{-}_i] +\\

&-2hf_i P^{+}_i \sum_{m=i+1}^{n} \frac{f_i}{f_m} \theta(\Delta f) \phi(\Delta f) \Delta v +\\ 

&+ P^{+}_i \sum_{m=0}^{i-1} \theta(-\Delta f) [P^{+}_i + P^{-}_i] + \\

&+ hf_i\sum_{m=0}^{i-1} \theta(-\Delta f) [ P^{+}_i + P^{-}_i] \phi(-\Delta f) \Delta v \\ 

\end{aligned}\\\\

&\begin{aligned}

F^{+}_i = &-\alpha_i + \eta_i \frac{P^{-}_i}{P^{+}_i} - \sum_{m=i+1}^{n} \frac{f_i}{f_m} \theta(\Delta f) [ P^{+}_i + P^{-}_i] +\\

&-2hf_i \sum_{m=i+1}^{n} \frac{f_i}{f_m} \theta(\Delta f) \phi(\Delta f) \Delta v +\\ 

&+ \sum_{m=0}^{i-1} \theta(-\Delta f) [P^{+}_i + P^{-}_i] + \\

&+ \frac{hf_i}{P^{+}_i}\sum_{m=0}^{i-1} \theta(-\Delta f) [ P^{+}_i + P^{-}_i] \phi(-\Delta f) \Delta v \\ 

\end{aligned}\\\\

&\begin{aligned}

F^{+}_i = &-\alpha_i + \eta_i \frac{P^{-}_i}{P^{+}_i} - \sum_{m=i+1}^{n} \frac{f_i}{f_m} \theta(\Delta f) \left[ [ P^{+}_i + P^{-}_i] + 2hf_i \phi(\Delta f) \Delta v \right] + \\ 

&+ \sum_{m=0}^{i-1} \theta(-\Delta f) [P^{+}_i + P^{-}_i] \left [1 + \frac{hf_i}{P^{+}_i} \phi(-\Delta f) \Delta v \right]\\ 

\end{aligned}

\end{aligned}

$$

## Backward
